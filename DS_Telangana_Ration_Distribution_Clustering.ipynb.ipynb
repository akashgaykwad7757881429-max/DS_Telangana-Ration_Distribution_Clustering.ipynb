```python
import pandas as pd

# Transaction files load
df1 = pd.read_csv("C:/Users/hp/Downloads/fpshop-card-status_1_2025.csv")
df2 = pd.read_csv("C:/Users/hp/Downloads/fpshop-card-status_2_2025.csv")
df3 = pd.read_csv("C:/Users/hp/Downloads/fpshop-card-status_3_2025.csv")
df4 = pd.read_csv("C:/Users/hp/Downloads/fpshop-card-status_4_2025.csv")
df5 = pd.read_csv("C:/Users/hp/Downloads/fpshop-card-status_5_2025.csv")

# Combine all 
trans_df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

print(trans_df.shape)
trans_df.head()
```

```python
location_df = pd.read_csv("C:/Users/hp/Downloads/shop-status-details_6_2025.csv")

print(location_df.shape)
location_df.head()
```

```python
# Card status files load
card1 = pd.read_csv("C:/Users/hp/Downloads/shop-wise-trans-details_1_2025.csv")
card2 = pd.read_csv("C:/Users/hp/Downloads/shop-wise-trans-details_2_2025.csv")
card3 = pd.read_csv("C:/Users/hp/Downloads/shop-wise-trans-details_3_2025.csv")
card4 = pd.read_csv("C:/Users/hp/Downloads/shop-wise-trans-details_4_2025.csv")
card5 = pd.read_csv("C:/Users/hp/Downloads/shop-wise-trans-details_5_2025.csv")

# Combine all card files
card_df = pd.concat([card1, card2, card3, card4, card5], ignore_index=True)

print(card_df.shape)
card_df.head()
```

```python
# Merge transaction + card status
merge1_df = pd.merge(
    trans_df,
    card_df,
    on=["shopNo", "month", "year"],
    how="left"
)

# Merge with location data
final_df = pd.merge(
    merge1_df,
    location_df,
    on="shopNo",
    how="left"
)

print(final_df.shape)
final_df.head()
```

```python
# Check missing values
print(final_df.isnull().sum())

# Check duplicate rows
print("Duplicate rows:", final_df.duplicated().sum())

# Data info
final_df.info()
```

```python
# Missing cvalues check 
print(final_df.isnull ().sum())

# Duplicate rows check
print("Duplicate rows:", final_df.duplicated().sum())
```

```python
# Missing values percentage check
missing_percent = (final_df.isnull().sum() / len(final_df)) * 100

print(missing_percent[missing_percent > 0].sort_values(ascending=False))
```

```python
# Fully empty columns remove
final_df.drop(
    columns=['cardTypeId', 'mroAsoApprDate', 'cardPoolType'],
    inplace=True
)

# Remaining missing values remove
final_df.dropna(inplace=True)

# Check shape after cleaning
print("Final Shape:", final_df.shape)

# Check missing values again
print(final_df.isnull().sum().sum())
```

```python
import matplotlib.pyplot as plt

# Top 10 districts by transactions
district_trans = final_df.groupby('distName_x')['totalUnits'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
district_trans.plot(kind='bar')
plt.title("Top 10 Districts by Total Units Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

features = [
    'utilization_ratio',
    'rice_wheat_ratio',
    'totalUnits',
    'totalRcs',
    'noOfTrans'
]

X = final_df[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled.shape)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)

final_df['Cluster'] = kmeans.fit_predict(X_scaled)

print(final_df[['Cluster']].head())
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)

final_df['PCA1'] = pca_result[:,0]
final_df['PCA2'] = pca_result[:,1]

plt.figure(figsize=(10,6))
plt.scatter(
    final_df['PCA1'],
    final_df['PCA2'],
    c=final_df['Cluster'],
    cmap='viridis',
    s=10
)

plt.title('PCA Visualization of Clusters')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=5)

final_df['DBSCAN_Cluster'] = dbscan.fit_predict(X_scaled)

print(final_df['DBSCAN_Cluster'].value_counts())
```

```python
cluster_profile = final_df.groupby('Cluster')[
    ['totalUnits','totalRcs','noOfTrans','utilization_ratio']
].mean()

print(cluster_profile)
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

st.set_page_config(page_title="Telangana PDS Analytics", layout="wide")

st.title("Telangana PDS Analytics Dashboard")

df = pd.read_csv("final_df.csv")

st.subheader("Dataset Overview")
st.write(df.head())

districts = st.sidebar.selectbox(
    "Select District",
    ["All"] + sorted(df["distName"].dropna().unique().tolist())
)

if districts != "All":
    df = df[df["distName"] == districts]

st.subheader("Cluster Distribution")
fig, ax = plt.subplots()
df["Cluster"].value_counts().sort_index().plot(kind="bar", ax=ax)
st.pyplot(fig)

st.subheader("Top 10 Shops by Total Units")
top_shops = df.groupby("shopNo")["totalUnits"].sum().nlargest(10)

fig2, ax2 = plt.subplots(figsize=(8,4))
top_shops.plot(kind="bar", ax=ax2)
st.pyplot(fig2)
```
